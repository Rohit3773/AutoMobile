#About Data

The dataset provides a comprehensive snapshot of various vehicles, encompassing their brand (Make), model (Model), and specific variant (Variant) details, along with their showroom prices (Ex-Showroom_Price). Technical specifications such as engine characteristics (Displacement, Cylinders, Valves_Per_Cylinder), drivetrain (Drivetrain), and fuel-related information (Fuel_System, Fuel_Tank_Capacity, Fuel_Type) are meticulously covered. Physical attributes like dimensions (Height, Length, Width), body type (Body_Type), and door count (Doors) are included. Additionally, the dataset offers insights into certified mileage (ARAI_Certified_Mileage), tire details (Front_Tyre_&Rim, Rear_Tyre&Rim), start/stop button availability (Start/Stop_Button), average fuel consumption (Average_Fuel_Consumption), multimedia options (CD/MP3/DVD_Player), safety features (ABS, EBD, Number_of_Airbags), and Electronic Brake Assist (EBA(Electronic_Brake_Assist)). This dataset serves as a valuable resource for gaining quick insights into a wide range of vehicle attributes.
